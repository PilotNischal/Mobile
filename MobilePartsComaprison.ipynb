```python
import matplotlib.pyplot as plt

# Sample data for internal memory (in GB) and battery power (in percentage)
int_memory = [2, 39, 36, 46, 45]
battery_power = [794, 1965, 1911, 1512, 510]

# Plotting the data
plt.figure(figsize=(10, 6))  # Set the figure size (width, height)
plt.plot(int_memory, battery_power, marker='o', linestyle='-', color='b', label='Battery Power (%)')

# Adding labels and title
plt.xlabel('Internal Memory (GB)')
plt.ylabel('Battery Power (%)')
plt.title('Battery Power vs. Internal Memory')

# Adding a legend
plt.legend()

# Adding grid lines
plt.grid(True)

# Show the plot
plt.tight_layout()  # Ensures all elements fit nicely in the plot area
plt.show()
```
